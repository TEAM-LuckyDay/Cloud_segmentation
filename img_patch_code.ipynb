```python
class image_processing():

    def __init__(self, img):
        self.image_size_5000 = self.crop_margin(img)
        self.make_patch_img(self.image_size_5000)

    

    def crop_margin(self,img):
        print("   : ", img.shape)
        self.crop_rgb_img = img[:,25:-25,:]
        print("   : ", self.crop_rgb_img.shape)
        self.image_5000 = self.crop_rgb_img[:5000,:5000,:]
        print("5000 size crop : " , self.image_5000.shape)

        return self.image_5000



    def make_patch_img(self, img):
        for i in range(5):
            for j in range(5):
                patch_img = img[ i*1000 : i*1000 + 1000,    j*1000 :  j*1000+1000  ,:]
                save_path = f'/content/drive/MyDrive/Cloud_data/cloud_train/train_clouds_data/patch_img/img{i+1}.jpg'
            
                cv2.imwrite(save_path, patch_img)
```
